### Part b.)

Since each patient has 50% chance of choosing each hospital --> 1000 coin flips

To solve this problem, we will be using the scipy binomial formula implementation; the documentation of which can be found [here](https://docs.scipy.org/doc/scipy/reference/tutorial/stats/discrete_binom.html?highlight=binomial).

The brute-force method:

````python
binom.pmf(1000, 1000, 0.5) # propbability everyone goes to one hosipital
````

````python
x=0
p_refusal= 0.01
p_choice = 0.5
N_patients = 1000
i = N_patients + 1
while x<p_refusal:
    i=i-1
    x=x+binom.pmf(i, N_patients, p_choice)

print('Both hospitals needs {} beds\n'.format(i));
````

#### plot distribution

````python
X=np.arange(0,N_patients)
Y = binom.pmf(X,N_patients,p_choice)
Y_c = binom.cdf(X,N_patients,p_choice)

fig, axs = plt.subplots(ncols=2, figsize=(10, 4))
axs[0].grid(True)
axs[0].plot(X,Y,'r',LineWidth=2); 
axs[0].set_title('Binomial PDF N={}, p={}'.format(N_patients,p_refusal));

axs[1].grid(True)
axs[1].plot(X,Y_c,'r',LineWidth=2);
axs[1].plot([0, 1000],[0.99, 0.99],'b--',LineWidth=2);
axs[1].set_title('Binomial CDF N={}, p={}'.format(N_patients,p_refusal)); # Cumulative distribution function
````

````python
# calculate the nice way

i_e = binom.ppf(0.99,N_patients,p_choice)
print('Both hospitals needs {} beds\n'.format(int(i_e)))
````

## Exercise 3
#### a.) we know that $\mu = 55$ and  `norm.cdf(50,$\mu$, $\sigma$)` = 0.1

````python
z = norm.ppf(0.1, 0, 1)
mu = 55
x = 50
sigma = (x - mu) / z
print(sigma)
````

#### let's check our result

````python
norm.cdf(x, mu, sigma)
````

#### b.) The precentage of cups that contain more than 61 ml:

````python
p = 1 - norm.cdf(61, mu, sigma)
print('~ {0:.0%} \n'.format(p))
````

#### c.) After adjustment, we have $\sigma = 3$ and `norm.cdf(50, $\mu$, $\sigma^2$) = 0.025`. Thus as in a.)

````python
sigma = 3
z = norm.ppf(0.025, 0, 1)
mu = x - z * sigma
print("{}ml per cup of coffee".format(mu))
````
